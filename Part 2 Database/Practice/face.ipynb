```python
from deepface import DeepFace
import numpy as np
import cv2
import matplotlib.pyplot as plt

```

```python
im_path1 = "images/yoona.jpg"
im_path2 = "images/yoona1.jpg"
img1 = cv2.imread(im_path1)
img2 = cv2.imread(im_path2)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 8))
plt.subplot(1, 2, 1)
plt.imshow(img1)
plt.subplot(1, 2, 2)
plt.imshow(img2)
plt.show()
```

```python
result = DeepFace.verify(img1_path = im_path1, img2_path = im_path2)
print(result)
```
